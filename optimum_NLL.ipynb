```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
l = [375424, 379316, 378060, 386811]
n = [8, 12, 16, 20]
plt.figure(figsize = (10, 7))
plt.plot(n, l, 's-', linewidth=3)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.xlabel('No. of clusters', fontsize = 20)
plt.ylabel('Negative log-likelihood', fontsize = 20)
plt.savefig('Negative log-likelihood.png', dpi=300, bbox_inches='tight');
```
